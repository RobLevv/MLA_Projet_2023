```python
import torch

from data_loader_V3 import ImgDataset, train_validation_test_split
from AutoEncoder import AutoEncoder
from Discriminator import Discriminator
from objectives import adversarial_objective, discriminator_objective

import matplotlib.pyplot as plt
import numpy as np
```

```python
ae = AutoEncoder()
ae.load_state_dict(torch.load("Models/autoencoder_e1.pt"))

discriminator = Discriminator()

dataset = ImgDataset(attributes_csv_file="data/Anno/list_attr_celeba.txt", img_root_dir="data/Img")
dataloader = torch.utils.data.DataLoader(dataset, batch_size=100, shuffle=True)

batch = dataloader.__iter__().__next__()
images = batch["image"]
attributes = batch["attributes"]

latent, decoded = ae(images, attributes)
print(latent.shape)
print(decoded.shape)
y_pred = discriminator(latent)

```

```python
print(discriminator)
```

```python
print("#### images ####")
print(np.max(images.detach().numpy(), axis=(0, 2, 3)))
print("#### decoded ####")
print(np.max(decoded.detach().numpy(), axis=(0, 2, 3)))
print("#### y_pred ####")
print(np.max(y_pred.detach().numpy()), np.min(y_pred.detach().numpy()))
print("#### attributes ####")
print(np.max(attributes.detach().numpy()), np.min(attributes.detach().numpy()))
print("#### 1-ypred ####")
print(np.max(1-y_pred.detach().numpy()), np.min(1-y_pred.detach().numpy()))
print("#### latent ####")
print(np.max(latent.detach().numpy()), np.min(latent.detach().numpy()))
```
